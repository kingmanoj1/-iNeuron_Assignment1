```python
from bs4 import BeautifulSoup 
import urllib.request 
import nltk
import requests
```

```python
url = "https://mlwhiz.com/blog/2020/10/30/mle-fb-interview/"
```

```python
response = requests.get(url)
```

```python
html_text = response.text
html_text
```

```python
soup = BeautifulSoup(html_text)
```

```python
soup.title.string
```

```python
text = " ".join([p.text for p in soup.find_all('p')])
```

```python
text
```

```python
import re
```

```python
tokens = re.findall('\w+', text)
```

```python
tokens = [token.lower() for token in tokens]
```

```python
nltk.download('stopwords')
sw = nltk.corpus.stopwords.words('english')
```

```python
len(sw)
```

```python
new_token = [word for word in tokens if word not in sw]
```

```python
new_token
```

```python
import collections
frequency = collections.Counter(new_token)
```

```python
frequency
```
